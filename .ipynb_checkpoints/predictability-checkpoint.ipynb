```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
from matplotlib import cm
import matplotlib.dates as mdates
import matplotlib.ticker as ticker
from matplotlib.dates import (YEARLY, MONTHLY, DateFormatter, MonthLocator,DayLocator,
                              rrulewrapper, RRuleLocator, drange)

from scipy.integrate import odeint
import numpy as np
from datetime import datetime
import seaborn as sns
%matplotlib inline
import arviz as az
import urllib.request
import pymc3 as pm
```

```python
path_montera34='https://github.com/montera34/escovid19data/blob/master/data/output/covid19-provincias-spain_consolidated.csv?raw=true'
```

```python
df = pd.read_csv(path_montera34)
```

```python
df['date']=pd.to_datetime(df['date'])
```

```python
fecha_inicio=datetime(year=2020,month=7,day=1)
provincias=['Madrid']
datos = df.fillna(0,axis=1)
datos = datos[(datos['date']>=fecha_inicio) & (datos['province'].isin(provincias))]

clave_avg='daily_cases_PCR_avg7'
clave_ratio_avg = 'ratio_daily_cases_PCR_avg7'
color_avg = 'darkred'
color_titulos = 'navy'
color_diarios = 'royalblue'
clave_ratio = 'tasa_ia_7_dias'
AI='incidencia_acumulada_14'
casos_PCR_14='cases_PCR_14days'
casos_nuevos = 'daily_cases_PCR'
clave_casos_popu = 'casos_per_population'
ventana_IA=14
titulo = 'Incidencia Acumulada de 14 días de PCR+ por 100.000 hab \n Ratio de la IA a 7 días'
y1='IA a 14 días'
y2='Ratio a 7 días'
  
for provincia in provincias:
    
    datos.loc[datos['province']==provincia,AI]=\
    datos.loc[datos['province']==provincia,casos_PCR_14]\
    /datos.loc[datos['province']==provincia,'poblacion']*100000

    datos.loc[datos['province']==provincia,clave_ratio] = \
    datos.loc[datos['province']==provincia][AI].pct_change(periods=7).add(1)    
    
    datos.loc[datos['province']==provincia,clave_casos_popu]=\
    datos.loc[datos['province']==provincia,casos_nuevos]\
    /datos.loc[datos['province']==provincia,'poblacion']*1000000
```

```python
casos_nuevos = np.array(datos['num_casos_prueba_pcr_avg7'].values[:-4])
casos_nuevos_log = np.log(casos_nuevos)
```

```python
t=np.array([i for i in range(len(casos_nuevos))])
```

```python
with pm.Model() as model:
    #t = pm.Data('tiempo',range(len(casos_nuevos)))
    t_=t
    beta = pm.Uniform('beta',lower=0,upper=1)
    r_mu = pm.Uniform('r_mu',lower=0,upper=1)
    #p = pm.Uniform('p',lower=0,upper=5)
    #q = pm.Uniform('q',lower=0,upper=5)
    sigma_I = pm.Gamma(name='sigmaI',alpha=0.01,beta=0.01)
    media = pm.Deterministic('media',(beta-r_mu)*t_)
    #log_I_obs = pm.Normal('log_I_obs',mu=beta/r_mu*p/(p+q),sigma=sigma_I,observed=casos_nuevos_log)
    log_I_obs = pm.Normal('log_I_obs',mu=media,sigma=sigma_I,observed=casos_nuevos_log)
    trace = pm.sample(10000, tune=1000,chains=1,cores=1)
                
    post_Pred = pm.sample_posterior_predictive(trace)
```

```python
with pm.Model() as model:
    t = pm.Data('tiempo',range(len(casos_nuevos)))
    
    beta = pm.Uniform('beta',lower=0,upper=1)
    r_mu = pm.Uniform('r_mu',lower=0,upper=1)
    p = pm.Uniform('p',lower=0,upper=5)
    q = pm.Uniform('q',lower=0,upper=5)
    
    sigma_I = pm.Gamma(name='sigmaI',alpha=0.01,beta=0.01)
    R0 = pm.Deterministic('R0',beta/r_mu)
    R0_ = pm.Deterministic('R0_',R0/(p+q)*(p+q*(1-np.exp(-(p+q)*t))/((p+q)*t)))
    media = pm.Deterministic('media',(R0_-1)*(beta-r_mu)*t)
    #log_I_obs = pm.Normal('log_I_obs',mu=beta/r_mu*p/(p+q),sigma=sigma_I,observed=casos_nuevos_log)
    log_I_obs = pm.Normal('log_I_obs',mu=media,sigma=sigma_I,observed=casos_nuevos_log)
    trace = pm.sample(10000, tune=1000,cores=1)
                
    post_Pred = pm.sample_posterior_predictive(trace)
```

```python
fig, ax = plt.subplots(figsize=(20, 8))
ax.plot(t, post_Pred['log_I_obs'][4].T, color="k", alpha=0.05)

ax.set_yscale("log")
ax.set(xlabel="Date", ylabel="Confirmed Cases", title='titulo')
plt.show()
```
