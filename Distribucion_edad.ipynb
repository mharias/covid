```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
from matplotlib import cm
import numpy as np
from datetime import datetime
import seaborn as sns
%matplotlib inline
```

```python
path='distri_edad.csv'
```

```python
df = pd.read_csv(path,encoding='ISO-8859-1',sep=';',engine='python',skipfooter=23)
```

```python
df.set_index(['Distribucion'],inplace=True)
```

```python
def format_y(value, tick_number):
    return '{:,.0f}'.format(value).replace(',','.')
```

```python
#f,(ax1,ax2,ax3,ax4) = plt.subplots(2, 2, figsize=(5, 10), sharex=True)
f = plt.figure(figsize=(10,5))
f.subplots_adjust(hspace=.5, wspace=.5)
#f.suptitle('titulo',fontsize=20,x=0,y=0)
ax1 = f.add_subplot(2,2,1)
# Generate some sequential data
sns.barplot(x=df.index, y=df.Confirmados, palette="deep", ax=ax1)
ax1.axhline(0, color="k", clip_on=False)
ax1.yaxis.set_major_formatter(plt.FuncFormatter(format_y))
y_axis = ax1.axes.get_yaxis().get_label().set_visible(False)
x_axis = ax1.axes.get_xaxis().get_label().set_visible(False)
ax1.text(0.02,.7,'Confirmados',fontsize=20,transform = ax1.transAxes)

# Center the data to make it diverging
ax2 = f.add_subplot(2,2,2)
sns.barplot(x=df.index, y=df.Hospitalizados, palette="deep", ax=ax2)
ax2.axhline(0, color="k", clip_on=False)
ax2.yaxis.set_major_formatter(plt.FuncFormatter(format_y))
y_axis = ax2.axes.get_yaxis().get_label().set_visible(False)
x_axis = ax2.axes.get_xaxis().get_label().set_visible(False)
ax2.text(0.02,.7,'Hospitalizados',fontsize=20,transform = ax2.transAxes)

# Randomly reorder the data to make it qualitative
ax3 = f.add_subplot(2,2,3)
sns.barplot(x=df.index, y=df.UCI, palette="deep", ax=ax3)
ax3.axhline(0, color="k", clip_on=False)
ax3.yaxis.set_major_formatter(plt.FuncFormatter(format_y))
y_axis = ax3.axes.get_yaxis().get_label().set_visible(False)
x_axis = ax3.axes.get_xaxis().get_label().set_visible(False)
ax3.text(0.02,.7,'UCI',fontsize=20,transform = ax3.transAxes)

ax4 = f.add_subplot(2,2,4)
sns.barplot(x=df.index, y=df.Fallecidos, palette="deep", ax=ax4)
ax4.axhline(0, color="k", clip_on=False)
ax4.yaxis.set_major_formatter(plt.FuncFormatter(format_y))
y_axis = ax4.axes.get_yaxis().get_label().set_visible(False)
x_axis = ax4.axes.get_xaxis().get_label().set_visible(False)
ax4.text(0.02,.7,'Fallecidos',fontsize=20,transform = ax4.transAxes)

# Finalize the plot
sns.despine(bottom=True)
#plt.setp(f.axes, yticks=[])
plt.tight_layout(h_pad=2)
```
